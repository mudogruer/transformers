```python
references = ["Today is a beautiful day.","Spaghetti is my favorite.","veracity."]
predictions = ["What a wonderful weather.","Albert Einstein was a German-born physicist .","accuracy ."]
```

```python
#evaluate with bert-score using distilbert model
from evaluate import load
import numpy as np
import pandas as pd

bertscore = load("bertscore")
predictions = predictions
references = references
results = bertscore.compute(predictions=predictions, references=references, model_type="distilbert-base-uncased")
d = {"references": references, "predictions":predictions, "Precision_Bert":results["precision"],"Recall_Bert":results["recall"],"F_Bert":results["f1"]}
df = pd.DataFrame(data=d)
df
```
