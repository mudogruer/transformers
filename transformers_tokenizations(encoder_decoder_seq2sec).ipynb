```python
text = "Human dignity shall be inviolable."
```

```python
from transformers import AutoTokenizer
```

```python
bert_tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
xlmr_tokenizer = AutoTokenizer.from_pretrained("FacebookAI/xlm-roberta-base")
gpt_tokenizer = AutoTokenizer.from_pretrained("openai-community/gpt2")
```

```python
bert_tokens = bert_tokenizer(text).tokens()
xlmr_tokens = xlmr_tokenizer(text).tokens()
gpt_tokens = gpt_tokenizer(text).tokens()
```

```python
import pandas as pd
```

```python
pd.DataFrame([bert_tokens,xlmr_tokens,gpt_tokens], index = ["BERT", "XLMR","GPT-2"])
```

```python
bert_tokens_ = bert_tokenizer(text)
xlmr_tokens_ = xlmr_tokenizer(text)
gpt_tokens_ = gpt_tokenizer(text)
xlmr_tokens_['input_ids']
pd.DataFrame([bert_tokens_['input_ids'],xlmr_tokens_['input_ids'],gpt_tokens_['input_ids']], index = ["BERT", "XLMR","GPT-2"])
```
